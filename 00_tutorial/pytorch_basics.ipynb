## 目的ディープラーニングフレームワークである PyTorch (`torch`) の基本的な使い方について学ぶ．

## PyTorchについて

PyTorch はディープラーニングのモデル構築・学習によく利用されるフレームワークです．
NumPy と似た書き方で多次元配列（Tensor）を扱うことができ，さらに

* GPU を利用した高速な演算
* 自動微分（autograd）による勾配計算

といった，ディープラーニングに必要な機能が組み込まれています．

以下では，PyTorch の基本的な使い方をプログラムを交えて説明します．

## PyTorchのインポート

PyTorch は `torch` という名前でインポートして使用します．

```python
import torch

print(torch.__version__)
```

## Tensor（テンソル）

PyTorch では，NumPy の `ndarray` に相当するデータ構造として `Tensor` を用います．
`torch.tensor()` にリストを渡すことで，Tensor を作成することができます．

```python
a = torch.tensor([[0, 0, 1], [0, 0, 2]], dtype=torch.float32)  # dtype=***でTensorの要素の型を指定します
b = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float32)
print(a)
print(b)
```

### Tensorの基本的な情報

`shape`（サイズ），`dtype`（要素の型），`device`（Tensorが存在する場所：CPU or GPU）を確認することができます．

```python
print(a.shape)
print(a.dtype)
print(a.device)
```

### NumPyとの相互変換

`torch.from_numpy()` で NumPy の `ndarray` から Tensor に変換することができます．
逆に，`.numpy()` で Tensor から `ndarray` に変換することができます．

（`ndarray` とTensorはメモリを共有するため，一方を変更するともう一方も変更される点に注意してください．）

```python
import numpy as np

np_array = np.array([[1, 2], [3, 4]], dtype=np.float32)

# ndarray -> Tensor
tensor_from_np = torch.from_numpy(np_array)
print(tensor_from_np)

# Tensor -> ndarray
array_from_tensor = tensor_from_np.numpy()
print(array_from_tensor)
```

## Tensorの演算

NumPy と同様に，Tensor 同士の四則演算や行列演算を行うことができます．

```python
# Tensorの演算（和・積）
c = a + b
d = a * b  # * は要素ごとの掛け算です
print(c)
print(d)

# （数学的な）行列の掛け算
e = torch.matmul(a, b.T)  # b.Tはbの転置を表しています
print(e)
```

## Tensorの扱い方

#### 特定の要素を取り出す（インデキシング）

`a[i, j]`：Tensor `a` の `(i, j)` 要素を取り出します．

#### 特定の範囲を切り出す（スライシング）

NumPy と同様，範囲の指定にはコロン「:」を用います．

`a[i, :]`：Tensor `a` の `i` 行目（のベクトル）

`a[:, j]`：Tensor `a` の `j` 列目（のベクトル）

```python
# インデキシング
a_elem = a[0, 2]
print(a_elem)

# スライシング
a_slice1 = a[1, :]
a_slice2 = a[:, 2]
print(a_slice1)
print(a_slice2)
```

#### Tensorの形状変更（reshape）

`view()` または `reshape()` を用いることで，Tensor の形状を変更することができます．
（`view()` はメモリを共有したまま形状のみを変更するのに対し，`reshape()` は必要に応じてデータをコピーします．）

```python
f = torch.arange(6, dtype=torch.float32)  # 0～5の要素を持つTensorを作成
print(f)

f_reshaped = f.view(2, 3)
print(f_reshaped)
```

## PyTorchの便利な関数

PyTorch にも NumPy と同様に，Tensor を操作するための関数が多数用意されています．
ここで全てを紹介することは難しいので，興味のある方は公式のreferenceページへアクセスして調べてみてください．

[PyTorch Documentation（公式ページ）](https://pytorch.org/docs/stable/index.html)

```python
g = torch.tensor([0, 1, 2], dtype=torch.float32)

# exp（指数関数を要素ごとに計算）
print(torch.exp(g))
# pow（累乗）
print(torch.pow(g, 2))

# 合計・平均などの集計関数
print(torch.sum(g))
print(torch.mean(g))
```

## GPUの利用

PyTorch では，Tensor を GPU 上に配置することで演算を高速化することができます．
`torch.cuda.is_available()` を用いることで，GPUが使用可能かどうかを確認することができます．

（Google Colab.でGPUを使用する場合は，メニューバーの「ランタイム」→「ランタイムのタイプを変更」からハードウェアアクセラレータをGPUにしてください．）

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

# Tensorを指定したデバイスへ転送
h = torch.tensor([1.0, 2.0, 3.0])
h = h.to(device)
print(h)
```

## 自動微分（autograd）

PyTorch には，Tensor の演算を記録し，自動的に勾配（微分値）を計算する「自動微分（autograd）」という機能があります．
ディープラーニングの学習（誤差逆伝播法）は，この機能を利用して実現されています．

Tensor 作成時に `requires_grad=True` を指定すると，その Tensor に対する演算が記録されるようになります．
`.backward()` を実行すると，記録された演算をもとに勾配が計算され，`.grad` に格納されます．

```python
x = torch.tensor(2.0, requires_grad=True)

# y = x^2 + 3x という計算を行う
y = x ** 2 + 3 * x

# yをxで微分（dy/dx = 2x + 3）
y.backward()

print(y)
print(x.grad)  # x=2.0のときの勾配（2*2+3=7）が表示されます
```
